La siguiente función la usarmeos para practicar con expresiones regulares. Podemos pasarle una expresión y un texto como parámetros y buscará la expresión en las líneas del texto, sacando por pantalla las líneas del texto en las que encuentra la expresión, y el núemro total de veces que la ha encontrado.

```python
def encontrar_patron(patron,documento):
  cuenta=0
  for i, linea in enumerate(documento):
    for match in re.finditer(patron, linea):
        cuenta=cuenta+1
        print ('Patrón encontrado en línea %s: %s' % (i+1, match.group()),'->', linea)
  print('Número de veces que se encuentra el patrón: ',cuenta)
```

La siguiente lista de patrones ejemplifica casos de expresiones regulares.

Usamos el método copile de re para convertir primero un objeto cadena mediante la que indicamos una expresión regular, a un objeto regexp con el que re trabaja. No es necesrio hacerlo así, podemos trabajar con la cadena directamente, sin compilarla previamente.

En algunas cadenas usamos el prefijo r para convertir la cadena en raw, y así evitar conflictos e significado entre el escape de caracteres mediante \ en python y en expresiones regulares, por ejemplo cuando usamos \b.

```python
#Patrones simples de Cadenas
patron = re.compile("guerra") #La cadena "guerra"
patron = re.compile("paz") #La cadena "paz"
patron = re.compile("guerra.+España") #La cadena "guerra" y después la cadena "España"
patron = re.compile("España.+guerra") #La cadena "España" y después la cadena "guerra"
patron = re.compile("(España|guerra).+(guerra|España)") #La cadena "España" y la cadena "guerra" en cualquier orden
patron = re.compile("[áéíóú]") #líneas que contienen vocales con tilde.
patron = re.compile("[ÁÉÍÓÚ]") #líneas que contienen vocales mayúsculas con tilde.
patron = re.compile("^[^a]+$") #líneas que no contienen la letra a
patron = re.compile("^[^A-Z]+$") #líneas que no contienen mayúsculas

#Patrones con inicio o fin de línea
patron = re.compile("^E") #líneas que empiezan con E
patron = re.compile("[A-Z]$") #líneas que acaban con una letra mayúscula

#Uso del prefijo r para hacer buen uso de la contrabarra
patron = re.compile("en") #la cadena "en"
patron = re.compile("\ben\b") #la palabra "en" pero no funciona
patron = re.compile(r"\ben\b") #la palabra "en"

patron = re.compile(r"\b[a-zA-Z]{4}\b") #palabras de cuatro letras
patron = re.compile(r"\b[a-zA-Z]{7,9}\b") #palabras de entre 7 y 9 letras


#Patrones con números.
patron = re.compile(r"\d+") #Una o más cifras seguidas
patron = re.compile(r"\d\d\d\d") #Cuatro cifras seguidas
patron = re.compile(r"\d\.\d\d\d") #Una cifra, un punto y tres cifras después (p.ej. fechas mal escritas)
patron= re.compile(r"\d+[^0-9]+\d+") #Dos números en la misma línea (pueden estar separados por un punto)
patron= re.compile(r"\d+[^0-9\.]+\d+") #Dos números en la misma línea (NO pueden estar separados por un punto)
patron = re.compile(r"^\n$") #lineas vacías (solo contienen el salto de línea)
```

patron = re.compile("^[abc]"): busca una cadena que comience con uno de los caracteres: a, b o c.

patron = re.compile("^[^a-zA-Z]"): buca cadenas que no comiencen con una letra, ya sea minúscula o mayúscula y de la A-Z y a-z.

patron = re.compile("^[^a-zA-Z\n]"): busca cadenas que no comiencen con ninguna letra ni con un salto de línea.

patron = re.compile("[abc]$"):busca cadenas que terminen con uno de los caracteres a, b o c.

patron = re.compile("^[^0-9]+$"): cadenas que consistan exclusivamente en caracteres que no sean números.

patron = re.compile(r"^[a-zA-Z]{4}\b"): cadenas que empiecen con exactamente 4 letras, ya sean mayúsculas o minúsculas y que luego tengan una frontera de palabra, es decir, el final de una palabra.
